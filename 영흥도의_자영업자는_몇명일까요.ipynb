```python
import folium
```

```python
# 위도
latitude = 37.2555
# 경도
longitude = 126.465726
```

```python
m = folium.Map(location=[latitude, longitude],
               zoom_start=13, 
               width=750, 
               height=500
              )
m
```

```python
folium.Marker([37.25804075645316, 126.49535631884356],
              popup="우리집",
              tooltip="우리집").add_to(m)
m
```

```python
folium.Marker([37.25804075645316, 126.49535631884356],
              popup="우리집",
              tooltip="우리집", 
              icon=folium.Icon('red', icon='star'),
             ).add_to(m)
m
```

```python
folium.CircleMarker([37.25804075645316, 126.49535631884356],
                    color='tomato',
                    radius = 50, 
                    tooltip='집주변 상권').add_to(m)
m
```

```python
import pandas as pd


df = pd.read_csv('/content/drive/MyDrive/소상공인시장진흥공단_상가(상권)정보_인천_202109.csv')
df.head(3)
```

```python
pd.Series(df.columns)
```

```python
sub_df = df.loc[df['행정동명'].isin(['영흥면'])]
sub_df.head(3)
```

```python
sub_df[['위도', '경도', '상호명']]
```

```python
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[37.25804075645316, 126.49535631884356],
    zoom_start=15
)

coords = sub_df[['위도', '경도']]


marker_cluster = MarkerCluster().add_to(m)

for lat, long in zip(coords['위도'], coords['경도']):
    folium.Marker([lat, long], icon = folium.Icon(color="green")).add_to(marker_cluster)
m
```
